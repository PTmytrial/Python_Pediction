```python
# import the data
from sklearn.datasets import load_iris
# loading the original data
original_data = load_iris()
# Multinomial Decision Tree - becase we are predicting flower type (no order)
print(original_data.target_names)
# These are the measurement that we are going to use
print(original_data.feature_names)
# extract the data portion only
X    = original_data.data
y    = original_data.target
print(y)
```

```python
# import pandas and convert data into DataFrame
import pandas as pd

X_df = pd.DataFrame(X)
X_df.columns = ["S Length","S Width","P Length","P Wide"]
X_df["Target"] = y

X_df
```

```python
# Train the DTC
from sklearn.tree import DecisionTreeClassifier

# to divide the dataset
from sklearn.model_selection import train_test_split

# Separate features and target
target   = X_df['Target']
features = X_df.drop('Target', axis=1)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)

# 150 * 20% = 30
print("traing data number : ",len(X_train))
print("testing data number : ",len(X_test))
```

```python
# Create and train the Decision Tree classsifier
ai = DecisionTreeClassifier()
ai.fit(X_train, y_train)

#Score the model
print(f"Model Accuracy on train data: {ai.score(X_train, y_train) * 100}%")
print(f"Model Accuracy on test data: {ai.score(X_test, y_test) * 100}%")
```

```python
# import necessary libraries
from sklearn.datasets        import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree            import DecisionTreeClassifier
from sklearn                 import tree
import matplotlib.pyplot as plt
import pandas as pd

# Load iris dataset
iris = load_iris()
X    = iris.data
y    = iris.target

X_df = pd.DataFrame(X)
X_df.columns = ["S Length","S Width","P Length","P Wide"]
X_df["Target"] = y

X_df_filterd = X_df[X_df["S Length"]>4.6]
```

```python
# prompt: Using dataframe X_df: I want to use Decision Tree classifier to Classify between flower types class label is the target column

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# Separate features and target
features = X_df.drop('Target', axis=1)
target = X_df['Target']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3)

# Create and train the Decision Tree classifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# Score the model
print(f"Model Accuracy: {clf.score(X_test, y_test) * 100}%")

# Visualize the decision tree
fig = plt.figure(figsize=(15,10))
_ = tree.plot_tree(clf,
                   feature_names=iris.feature_names,
                   class_names=iris.target_names,
                   filled=True)
```
